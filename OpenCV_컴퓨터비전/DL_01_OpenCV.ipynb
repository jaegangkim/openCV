### 이진( Binary) 이미지 만들기
- 픽셀값 0 ~ 255

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./image/10.png', cv2.IMREAD_GRAYSCALE)

# threshold : 한계점
ret, thr1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thr1, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

- cv2.threshold(img,threshold_value, value, flag)
    - threshold_value : 픽셀 경계값 ( 0과 255로 나누기위한 기준 )
    - value : 경계값보다 클 때 적용되는 값
    - flag : 적용 방법
        - cv2.THRESH_BINARY : 픽셀 값이 경계깞보다 크면 value, 아니면 0을 할당
        - cv2.THRESH_BINARY_INV : 반대
        - cv2.THRESH_TRUNC : 픽셀 값이 경계값보다 크면 경계값, 아니면 픽셀 값 그대로 할당
        - cv2.THRESH_TOZERO :      ''            크면 픽셀값 아니면 0
        - cv2.THRESH_TOZERO_INV : 반대

### 도형 그리기

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./image/10.png', cv2.IMREAD_COLOR)

# img = cv2.rectangle(img, (10,10), (150,150), (0,0,255))
img = cv2.putText(img, "only Eng", (10,50),
                  cv2.FONT_HERSHEY_SIMPLEX,
                  1, (255,0,0), 2 )

cv2.imshow('IU', img)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

- cv2.line(배경이미지, 시작점좌표, 끝점좌표, 색상, 선두께) : 직선 그리기 함수
- cv2.circle(배경이미지, 중심좌표, 반지름, 색상, 선두께) : 원 그리기 함수
- cv2.rectangle(배경이미지, 좌상단점좌표, 우하단점좌표, 색상, 선두께) : 사각형 그리기 함수
- cv2.ellipse(배경이미지, 중심좌표, 장축/단축길이, 호의시작각, 호의 종료각, 색상, 선두께) : 타원 그리기 함수
- cv2.putText(배경이미지, 출력내용, 출력시작좌표, 폰트, 크기, 색상, 굵기) : 텍스트 입력 함수

### 한글 출력하기

```python
from PIL import ImageFont, ImageDraw, Image
import numpy as np
import cv2
```

```python
img = cv2.imread('./image/10.png', cv2.IMREAD_COLOR)

# 폰트와 폰트 크기 설정
font = ImageFont.truetype('fonts/gulim.ttc', 30)

# 이미지를 PIL 이미지로 변환
img_pil = Image.fromarray(img)
draw = ImageDraw.Draw(img_pil)

# fill= ( b,g,r,alpha )
draw.text((10,20), '한글 가즈아', font = font, fill = (0,0,0,255))
img = np.array(img_pil)

cv2.imshow('iu',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
