### 색상공간 (Color Space) 이해하기
- RGB : Red, Green, Blue로 구성된 색상공간 - PC의 모니터에 주로 사용
    - 가산혼합 : 색을 섞을수록 명도가 높아지는 혼합

- CMYK : Cyan, Magenta, Yellow, Black으로 구성된 색상공간 - 인쇄용으로 주로 사용
    - 감산혼합 : 색을 섞을수록 명도가 낮아지는 혼합

- HSV : Hue(색상), Saturation(채도), Value(명도)로 변환한 색상공간
    - 채도 : 색의 진하고 연한 정도
    - 명도 : 밝기
    - 직관적으로 색상 파악하기에 용이

- 색상 값
  - Hue 값은 0-179 범위로 설정
  - 만약 Hue 값을 -로 설정하면 뒤쪽부터 인덱싱됨

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./images/balloon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
red = np.uint8([[[0,0,255]]])
hsv_red = cv2.cvtColor(red, cv2.COLOR_BGR2HSV)
print(hsv_red)
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)

# HSV 색상공간에서 범위값에 해당하는 부분 추출
# 비슷한 색깔은 h-10, 100, 100 에서 h+10, 255, 255 사이
mask = cv2.inRange(hsv, (100,100,100),(130,255,255)) #내가 찾고자 하는 색상을 추적하는 코드

img2 = cv2.bitwise_and(img, img, mask = mask)

plt.imshow(img2)
```

```python
m = cv2.imread('./images/candies.png')
m = cv2.cvtColor(m, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(m,cv2.COLOR_RGB2HSV)

# HSV 색상공간에서 범위값에 해당하는 부분 추출
# 비슷한 색깔은 h-10, 100, 100 에서 h+10, 255, 255 사이
mask2 = cv2.inRange(hsv, (47,100,100),(74,255,255)) #내가 찾고자 하는 색상을 추적하는 코드

img3 = cv2.bitwise_and(m, m, mask = mask2)

plt.imshow(img3)
```

### 크로마 키 (Chroma key) 합성
- 녹색 또는 파란색 배경에서 촬영한 영상에 다른 배경 영상을 합성하는 기술

```python
img = cv2.imread('./images/chromakey2.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

back = cv2.imread('./images/back2.png')
back = cv2.cvtColor(back, cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.imshow(img)

plt.subplot(1,2,2)
plt.imshow(back)

plt.show()
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

mask = cv2.inRange(hsv, (50,100,100), (70,255,255))
```

```python
cv2.copyTo(back, mask, img)
# back , mask : mask가 0이 아닌 픽셀에 대해서 back의 이미지 복사연산 수행

plt.imshow(img)
```

### 이미지 필터링
- blur (노이즈 제거하는데 유용)

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./images/tiger/lenna.png')

def plotFigure(o_image, b_image) :
    plt.figure(figsize=(10,10))
    
    plt.subplot(1,2,1)
    plt.title('Original')
    plt.imshow(cv2.cvtColor(o_image,cv2.COLOR_BGR2RGB))
    
    plt.subplot(1,2,2)
    plt.title('Blur')
    plt.imshow(cv2.cvtColor(b_image,cv2.COLOR_BGR2RGB))
    
    plt.show()
```

#### averaging(필터 박스 내의 평균을 이용한 방법)
- 필터 박스 안의 픽셀들의 값을 평균을 내어 대체한다.

```python
blur1 = cv2.blur(img, (5,5)) # 5,5 크기의 필터
plotFigure(img, blur1)
```

#### Gaussian Blur
- 가우시안 분포를 이용하여 블러처리 하는 방식
- 가우시안 분포의 가중치로 평균을 내어 픽셀값을 업데이트
- 가중치는 중심 픽셀에서 멀어질수록 줄어든다.
- 중심 픽셀의 값은 그 주변의 픽셀들의 영향을 가장 크게 받는다.

```python
blur2 = cv2.GaussianBlur(img, (15,15),0) # 표준편차
# 현재 픽셀에만 가중치가 곱해진다.

plotFigure(img, blur2)
```

#### Median Blur(중앙값)
- 현재 픽셀값을 필터내의 픽셀값의 중앙값으로 대체한다.
- 노이즈 데이터에 대한 영향이 적다

```python
blur3 = cv2.medianBlur(img, 7)
plotFigure(img,blur3)
```

#### Bilateral Filter(바이레터럴 필터)
- 지금까지 적용한 블러는 노이즈를 제거하는 효과는 뛰어나지만 그만큼 이미지 객체간의 경계도 흐릿하게 만드는 문제가 있다.
- 가우시안 필터와 경계 필터를 결합하여 만들어진 필터

```python
# (img, d, sigmaColor, sigmaSpace)
# d : 필터크기
#
blur4 = cv2.bilateralFilter(img,10,75,75)

plotFigure(img,blur4)
```

### 모폴로지 - 침식과 팽창
- morphology : 형태학
- 노이즈 제거 , 구멍 채우기, 끊어진 선 이어 붙이기 등 형태학적인 연산을 말한다.
- 모폴로지 연산은 검은색과 흰색으로만 구성되어있는 바이너리(binary) 이미지에 적용할 수 있다.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./images/number.png', cv2.IMREAD_GRAYSCALE)
```

```python
def plotFigure2(o_image, b_image):
    plt.figure(figsize=(10,10))
    plt.subplot(1,2,1)
    plt.title('Original')
    plt.imshow(cv2.cvtColor(o_image, cv2.COLOR_BGR2RGB))
    
    plt.subplot(1,2,2)
    plt.title('morphology') 
    plt.imshow(cv2.cvtColor(b_image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
# 필터 크기 설정
kernel = np.ones((3,3),np.uint8)

erosion = cv2.erode(img, kernel, iterations =1) #침식


dilation = cv2.dilate(img, kernel,iterations=1) # 팽창
```

```python
plotFigure2(erosion,dilation)
```

```python
a = cv2.imread('./images/noise_A.png')
a_erosion = cv2.erode(a,kernel, iterations=1)

b = cv2.imread('./images/noise_B.png')
b_dilation = cv2.dilate(b, kernel, iterations=1)
plotFigure2(a_erosion,b_dilation)
```

### Opening과 Closing
- Opening: erosion 수행 후에 바로 dilate를 수행하여 원래 이미지 크기로 돌려 놓는다.
- Closing : dilate 수행 후에 바로 erosion을 수행하여 원래 이미지 크기로 돌려 놓는다.

```python
opening = cv2.morphologyEx(a, cv2.MORPH_OPEN, kernel, iterations=1)

closing = cv2.morphologyEx(b,cv2.MORPH_CLOSE, kernel, iterations=1)

plotFigure2(opening,closing)
```

```python
obj = cv2.imread('./images/object.png', cv2.IMREAD_GRAYSCALE)
opeing_obj = cv2.morphologyEx(obj, cv2.MORPH_OPEN, kernel, iterations=3)
plt.imshow(opeing_obj, cmap='gray')
```
