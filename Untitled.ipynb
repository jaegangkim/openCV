### 크롤링으로 수집한 이미지 데이터 전처리 작업

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
# os : 파일, 폴더 처리
import os
```

```python
pica_dir = '피카츄/'
hamster_dir = "햄스터/"
rai_dir = "라이츄/"
```

```python
# 파일명들을 리스트에 저장 해줌
pica_fnames = os.listdir(pica_dir)
hamster_fnames = os.listdir(hamster_dir)
rai_fnames = os.listdir(rai_dir)
len(pica_fnames), len(hamster_fnames), len(rai_fnames)
```

```python
def load_images(folder_path, file_names, img_size_shape=(224,224)):
    images = []
    for i in file_names:
        # 폴더 경로와 파일 이름 결합
        path = os.path.join(folder_path, i)
        # 파일 오픈 및 크기조정
        img = Image.open(path).resize(img_size_shape).convert('RGB')
        # numpy배열로 변경 + 리스트 추가
        images.append(np.array(img))
    # 리스트도 numpy배열로 변환하여 리턴
    return np.array(images)
```

```python
train_pica = load_images(pica_dir, pica_fnames)
train_hamster = load_images(hamster_dir, hamster_fnames)
train_rai = load_images(rai_dir, rai_fnames)
```

```python
train_pica.shape, train_hamster.shape, train_rai.shape # 3의 의미는 RGB값
```

```python
plt.imshow(train_pica[0])
```

```python
# 문제데이터 생성
X = np.concatenate([train_pica, train_hamster, train_rai])
```

```python
X.shape
```

```python
# 정답데이터 생성
# 0 : 피카츄 , 1: 햄스터, 2: 라이츄
y = np.array([0]*412 + [1]*400 + [2]*400)
```

```python
# 학습, 평가 데이터로 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=0.2 ,
                                                   random_state=543)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

- NPZ(Numpy Zip) : numpy 데이터를 압축시켜 놓은 형태

```python
np.savez_compressed('animals.npz',
                   X_train = X_train,
                    X_test=X_test,
                    y_train=y_train,
                    y_test=y_test
                   )
```
